```python
missing_percentage = data.isnull().mean() * 100
missing_percentage
if(missing_percentage > 0).any():
   print(missing_percentage[missing_percentage > 0])
   print("There are missing values in the dataset.")
else:
   print("There are no missing values in the dataset.")
```

```python
data.head(5)
```

```python
unknown_data = data[
    (data['job'] == 'unknown') |
                    (data['marital'] == 'unknown') |
                      (data['education'] == 'unknown') |
                       (data['default'] == 'unknown') |
                        (data['housing'] == 'unknown') |
                         (data['loan'] == 'unknown')
]

unknown_data.head(5)
```

```python
data.replace('unknown', pd.NA, inplace=True)
data.dropna(inplace=True)
```

```python
check_unknown = data[
    (data['job'] == 'unknown') |
                    (data['marital'] == 'unknown') |
                      (data['education'] == 'unknown') |
                       (data['default'] == 'unknown') |
                        (data['housing'] == 'unknown') |
                         (data['loan'] == 'unknown')

]
if(check_unknown.empty):
   print("There are no unknown values in the dataset.")
else:
   print("There are unknown values in the dataset.")
```

```python
data['campaign']
```

```python
data['contact_efficiency'] = data['duration'] / data['campaign']
```

```python
data.head(5)
```

```python
data.info()
```

```python
data['is_repeat_customer'] = data['pdays'].apply(lambda x: 1 if x != 999 else 0)
```

```python
data.head(5)
```

```python
sns.histplot(data['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.barplot(x='job', y='y', data=data)
plt.xticks(rotation=45)
plt.title('Conversion Rate by Job')
plt.show()
```
